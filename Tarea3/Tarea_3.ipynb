```python
import pandas as pd  # Librería para trabajar con datos en tablas
```

```python
clientes = pd.read_csv("clientes.csv")
```

```python
ventas = pd.read_csv("ventas.csv")
```

```python



categorias = pd.read_csv("categorias.csv")
productos = pd.read_csv("productos.csv")

detalle_ventas = pd.read_csv("detalle_ventas.csv")

# 2️⃣ Revisar los datos cargados (primeras filas de cada tabla)
print("Clientes:")
print(clientes.head(), "\n")

print("Categorías:")
print(categorias.head(), "\n")

print("Productos:")
print(productos.head(), "\n")

print("Ventas:")
print(ventas.head(), "\n")

print("Detalle de Ventas:")
print(detalle_ventas.head())
```

```python
print("Clientes:", clientes.columns)
print("Categorías:", categorias.columns)
print("Productos:", productos.columns)
print("Ventas:", ventas.columns)
print("Detalle ventas:", detalle_ventas.columns)
```

```python
# Ajustar tipos de datos usando los nombres EXACTOS de las columnas
clientes = clientes.astype({
    "id cliente": "int",
    "nombre cliente": "string",
    "telefono": "string",
    "correo": "string"
})

categorias = categorias.astype({
    "Id categoria": "int",
    "Nombre": "string"
})

productos = productos.astype({
    "Id producto": "int",
    "Nombre producto": "string",
    "Descripcion": "string",
    "Id Categoria": "int",
    "Precio": "float",
    "stock": "int"
})

ventas = ventas.astype({
    "id_venta": "int",
    "fecha_compra": "string",
    "id_cliente": "int",
    "cantidad_compra": "int"
})

detalle_ventas = detalle_ventas.astype({
    "id_detalle": "int",
    "id_venta": "int",
    "id_producto": "int",
    "precio_unitario": "float",
    "cantidad_compra": "int"
})

# Opcional: usar id_detalle como índice en detalle_ventas
detalle_ventas.set_index("id_detalle", inplace=True)

print("✅ Tipos de datos ajustados correctamente")
```

```python
# --- Copias con nombres estandarizados (snake_case) ---
cli = clientes.rename(columns={
    "id cliente":"id_cliente",
    "nombre cliente":"nombre_cliente",
    "telefono":"telefono",
    "correo":"correo"
})

cat = categorias.rename(columns={
    "Id categoria":"id_categoria",
    "Nombre":"nombre_categoria"
})

prod = productos.rename(columns={
    "Id producto":"id_producto",
    "Nombre producto":"nombre_producto",
    "Descripcion":"descripcion",
    "Id Categoria":"id_categoria",
    "Precio":"precio_lista",
    "stock":"stock"
})

ven = ventas.rename(columns={
    "id_venta":"id_venta",
    "fecha_compra":"fecha_compra",
    "id_cliente":"id_cliente",
    "cantidad_compra":"cantidad_compra_venta"  # si no lo usas, se puede ignorar
})

det = detalle_ventas.rename(columns={
    "id_venta":"id_venta",
    "id_producto":"id_producto",
    "precio_unitario":"precio_unitario",
    "cantidad_compra":"cantidad"
})

# Convertir fecha a datetime (útil para agrupar por mes)
ven["fecha_compra"] = pd.to_datetime(ven["fecha_compra"], errors="coerce", dayfirst=False)

print("✅ Columnas estandarizadas y fecha convertida.")
```

```python
# Crear la tabla ventas_completas uniendo todo
ventas_completas = (
    det
    .merge(prod, on="id_producto", how="left")
    .merge(ven, on="id_venta", how="left")
    .merge(cli, on="id_cliente", how="left")
    .merge(cat, on="id_categoria", how="left")
)

# Crear la columna total_linea
ventas_completas["total_linea"] = ventas_completas["cantidad"] * ventas_completas["precio_unitario"]

# Verifica que se creó
ventas_completas.head()
```

```python
ventas_completas = ventas_completas.set_index("id_venta")
ventas_completas.head()
```

```python
# Crear una nueva columna: IVA y Total con IVA
ventas_completas["iva"] = ventas_completas["total_linea"] * 0.16
ventas_completas["total_con_iva"] = ventas_completas["total_linea"] + ventas_completas["iva"]

# Modificar columna: redondear total_con_iva a 2 decimales
ventas_completas["total_con_iva"] = ventas_completas["total_con_iva"].round(2)

# Eliminar una columna (ejemplo quitar iva)
# ventas_completas.drop(columns=["iva"], inplace=True)

ventas_completas.head()
```

```python
# Filtrar ventas mayores a $1000
ventas_mayores = ventas_completas[ventas_completas["total_linea"] > 1000]
ventas_mayores.head()
```

```python
# Subconjunto con columnas clave
subconjunto = ventas_completas[["nombre_cliente", "nombre_producto", "cantidad", "total_con_iva"]]
subconjunto.head()
```

```python
# Total vendido por cliente
total_por_cliente = ventas_completas.groupby("nombre_cliente")["total_linea"].sum()

# Total vendido por categoría
total_por_categoria = ventas_completas.groupby("nombre_categoria")["total_linea"].sum()

print("Total por cliente:\n", total_por_cliente)
print("\nTotal por categoría:\n", total_por_categoria)
```
