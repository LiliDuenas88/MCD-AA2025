**METODOLOGÍA**

La metodología se estructuró en cinco fases integradas, cada una sustentada en herramientas estadísticas y computacionales robustas.

**1. Análisis descriptivo y pruebas de normalidad**

Se utilizaron estadísticas básicas (media, mediana, desviación estándar, cuartiles) y visualizaciones (histogramas, boxplots, matriz de correlación) para caracterizar las variables: precio unitario, cantidad compra, venta total y stock.
Posteriormente se aplicaron:

•	Shapiro–Wilk para normalidad univariada.

•	Distancia de Mahalanobis + QQ plot.

•	Prueba multivariada de Mardia para evaluar la normalidad conjunta.

El análisis mostró que ninguna variable es normal y que existe fuerte asimetría multivariada.

**2. Selección de características**

Se evaluó la relevancia de precio unitario, cantidad compra y venta total mediante cuatro técnicas:
•	F-Test (ANOVA)

•	Mutual Information

•	Random Forest Importance

•	Lasso (L1)

Esto permitió determinar qué variables contribuyen significativamente a predecir el stock. Todos los métodos coincidieron en que precio unitario es el factor dominante.

**3. Agrupamiento no supervisado**

Se estandarizaron las variables y se aplicó DBSCAN, justificado porque:

•	No requiere especificar número de grupos.

•	Identifica estructuras arbitrarias y outliers.

•	Funciona bien con datos ruidosos y densidad variable.

La calidad de los agrupamientos se evaluó con tres índices:

•	Silhouette

•	Calinski–Harabasz

•	Davies–Bouldin

Los valores obtenidos indicaron grupos válidos y estructura diferenciada en los datos.

**4. Modelo de predicción supervisado**

Se construyó un modelo Random Forest Regressor, ideal para datos no lineales y no normales. Se entrenó y evaluó con las métricas:

•	MAE

•	RMSE

•	R²

El R² = 0.9995 evidenció un ajuste sobresaliente.
Se validó la importancia de variables y se realizaron predicciones para nuevos escenarios simulados.

**5. Diseño de Experimentos (DOE 2³)**

Se aplicó un diseño factorial 2³ para evaluar los efectos principales de:

•	Precio unitario (P)

•	Cantidad de compra (C)

•	Venta total (V)

Así como interacciones P×C, P×V, C×V y P×C×V.
El DOE se ejecutó utilizando como respuesta el stock predicho por el Random Forest, integrando así el modelo con una metodología experimental clásica.
